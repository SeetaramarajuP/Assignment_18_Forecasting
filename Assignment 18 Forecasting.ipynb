# Coco_Cola Sales Question 1

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
cc=pd.read_excel('CocaCola_Sales_Rawdata.xlsx')
cc
```

```python
cc.plot()
plt.show()
```

```python
cc.hist()
plt.show()
```

```python
cc.plot(kind='kde')
plt.show()
```

```python
from pandas.plotting import lag_plot
lag_plot(cc['Sales'])
plt.show()
```

```python
len(cc)
```

```python
cc['quarter'] = 0
for i in range(42):
    p=cc['Quarter'][i]
    cc['quarter'][i]=p[0:2]
cc
```

```python
cc_dummies=pd.get_dummies(cc['quarter'])
cc_dummies
```

```python
cc=pd.concat([cc,cc_dummies],axis= 1)
cc
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose_ts_add=seasonal_decompose(cc.Sales, period=12)
decompose_ts_add.plot()
plt.show()
```

```python
cc['t'] = np.arange(1,43)
cc['t_squared'] = cc['t']**2
cc["Sales_log"] =np.log(cc['Sales'])
```

```python
train=cc.head(30)
test=cc.tail(12)
```

```python
import statsmodels.formula.api as smf
m_linear=smf.ols('Sales~t',data=train).fit()
pred_linear=pd.Series(m_linear.predict(test['t']))
rmse_linear=np.sqrt(np.mean(np.array(test['Sales'])-np.array(pred_linear))**2)
rmse_linear
```

```python
m_exp=smf.ols('Sales_log~t',data=train).fit()
pred_exp=pd.Series(m_exp.predict(test['t']))
rmse_exp=np.sqrt(np.mean(np.array(test['Sales'])-np.array(pred_exp))**2)
rmse_exp
```

```python
m_quad=smf.ols('Sales~t+t_squared',data=train).fit()
pred_quad=pd.Series(m_quad.predict(test[['t','t_squared']]))
rmse_quad=np.sqrt(np.mean(np.array(test['Sales'])-np.array(pred_quad))**2)
rmse_quad
```

```python
add_sea=smf.ols('Sales~Q1+Q2+Q3+Q4',data=train).fit()
pred_add_sea=pd.Series(add_sea.predict(test[['Q1','Q2','Q3','Q4']]))
rmse_add_sea=np.sqrt(np.mean(np.array(test['Sales'])-np.array(pred_add_sea))**2)
rmse_add_sea
```

```python
data = {"MODEL":pd.Series(["rmse_linear","rmse_exp","rmse_quad","rmse_add"]),"RMSE_Values":pd.Series([rmse_linear,rmse_exp,rmse_quad,rmse_add_sea])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from statsmodels.tsa.holtwinters import Holt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
c_c=pd.read_excel('CocaCola_Sales_Rawdata.xlsx')
c_c
train=c_c.head(30)
test=c_c.tail(12)
```

```python
plt.figure(figsize=(12,4))
c_c.Sales.plot(label='org')
for i in range(2,24,6):
    cc['Sales'].rolling(i).mean().plot(label=str(i))
plt.legend(loc='best')
```

```python
train['Sales']=train['Sales'].astype('double')
ses_model=SimpleExpSmoothing(train['Sales']).fit(smoothing_level=0.2)
pred_ses=ses_model.predict(start=test.index[0],end=test.index[-1])
rmse_ses=np.sqrt(np.mean(np.array(test['Sales'])-np.array(pred_ses))**2)
rmse_ses
```

```python
holt_model=Holt(train['Sales']).fit(smoothing_level=0.2,smoothing_slope=0.15)
pred_holt=holt_model.predict(start=test.index[0],end=test.index[-1])
rmse_holt=np.sqrt(np.mean(np.array(test['Sales'])-np.array(pred_holt))**2)
rmse_holt
```

```python
hw_model=ExponentialSmoothing(train['Sales'],seasonal='add',trend='add',seasonal_periods=12).fit()
pred_hw=hw_model.predict(start=test.index[0],end=test.index[-1])
rmse_hw=np.sqrt(np.mean(np.array(test['Sales'])-np.array(pred_hw))**2)
rmse_hw
```

```python
data_ = {"MODEL":pd.Series(["rmse_ses","rmse_holt","rmse_hw"]),"RMSE_Values":pd.Series([rmse_ses,rmse_holt,rmse_hw,])}
table_rmse=pd.DataFrame(data_)
table_rmse.sort_values(['RMSE_Values'])
```

RSME Value of Holt_ Winter is tge least
